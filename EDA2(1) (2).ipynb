```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
from sklearn.ensemble import IsolationForest

# Load the dataset
data = pd.read_csv('adult_with_headers.csv')

# Assuming ' <=50K' is your target variable
data['income'] = data['income'].apply(lambda x: 0 if x.strip() == '<=50K' else 1)

# Handle missing values
num_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='most_frequent')

# Separate numerical and categorical columns
num_cols = data.select_dtypes(include=['int64', 'float64']).columns
cat_cols = data.select_dtypes(include=['object']).columns

# Handle numerical columns
data[num_cols] = num_imputer.fit_transform(data[num_cols])

# Handle categorical columns
data[cat_cols] = cat_imputer.fit_transform(data[cat_cols])

# One-hot encode categorical columns
encoder = OneHotEncoder(drop='first', sparse=False)
encoded_cats = encoder.fit_transform(data[cat_cols])
encoded_cols = encoder.get_feature_names_out(cat_cols)  # Get feature names after encoding
encoded_cats_df = pd.DataFrame(encoded_cats, columns=encoded_cols)
data_encoded = pd.concat([data[num_cols], encoded_cats_df], axis=1)

# Create new features
data_encoded['age_squared'] = data_encoded['age'] ** 2
data_encoded['hours_per_week_log'] = np.log1p(data_encoded['hours_per_week'])

# Check skewness of numerical features
skewness = data_encoded[num_cols].apply(lambda x: x.skew())

# Apply log transformation to skewed numerical features
skewed_cols = skewness[skewness > 1].index
data_encoded[skewed_cols] = data_encoded[skewed_cols].apply(lambda x: np.log1p(x))

# Use Isolation Forest to identify and remove outliers
isolation_forest = IsolationForest(contamination=0.05, random_state=42)
outliers = isolation_forest.fit_predict(data_encoded)
data_if = data_encoded[outliers != -1]

# Calculate the correlation matrix
correlation_matrix = data_if.corr()

correlation_matrix
```
