```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset
file_path = 'anime.csv'  # Update this with the actual path to your CSV file
data = pd.read_csv(file_path)

# Check for missing values and handle them
data.fillna('', inplace=True)

# Encode categorical features if necessary
label_encoder = LabelEncoder()
data['genre_encoded'] = label_encoder.fit_transform(data['genre'])  # Example if 'genre' is a categorical feature

# Normalize numerical features if required
numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns
scaler = StandardScaler()
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])

# Compute cosine similarity matrix for a feature set
features = ['genre_encoded']  # Update with relevant feature names
similarity_matrix = cosine_similarity(data[features])

# Define a function to recommend anime based on cosine similarity
def recommend_anime(anime_title, data, similarity_matrix, top_n=10):
    if anime_title not in data['anime_id'].values:
        return "Anime not found in the dataset"
    
    # Get the index of the anime that matches the title
    idx = data[data['anime_id'] == anime_title].index[0]

    # Get the pairwsie similarity scores of all anime with that anime
    sim_scores = list(enumerate(similarity_matrix[idx]))

    # Sort the anime based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the top n most similar anime
    sim_scores = sim_scores[1:top_n+1]

    # Get the anime indices
    anime_indices = [i[0] for i in sim_scores]

    # Return the top n most similar anime
    return data['anime_id'].iloc[anime_indices]

# Example: Get recommendations for a specific anime
recommendations = recommend_anime('Naruto', data, similarity_matrix)  # Replace 'Naruto' with the title of an anime in your dataset
print(recommendations)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset
  
data = pd.read_csv('anime.csv')

# Handle missing values
data.fillna('', inplace=True)

# Encode categorical features if necessary
label_encoder = LabelEncoder()
data['genre_encoded'] = label_encoder.fit_transform(data['genre'])  # Example if 'genre' is a categorical feature

# Normalize numerical features if required
numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns
scaler = StandardScaler()
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])

# Split the dataset into training and testing sets
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

# Compute cosine similarity matrix for a feature set
features = ['genre_encoded']  # Update with relevant feature names
similarity_matrix = cosine_similarity(data[features])

# Define a function to recommend anime based on cosine similarity
def recommend_anime(anime_title, data, similarity_matrix, top_n=10):
    if anime_title not in data['anime_id'].values:
        return "Anime not found in the dataset"
    
    # Get the index of the anime that matches the title
    idx = data[data['anime_id'] == anime_title].index[0]

    # Get the pairwsie similarity scores of all anime with that anime
    sim_scores = list(enumerate(similarity_matrix[idx]))

    # Sort the anime based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the top n most similar anime
    sim_scores = sim_scores[1:top_n+1]

    # Get the anime indices
    anime_indices = [i[0] for i in sim_scores]

    # Return the top n most similar anime
    return data['anime_id'].iloc[anime_indices]

# Evaluate the recommendation system using appropriate metrics
def evaluate_recommendations(test_data, similarity_matrix, top_n=10):
    y_true = []
    y_pred = []

    for anime_title in test_data['anime_id']:
        recommendations = recommend_anime(anime_title, data, similarity_matrix, top_n)
        y_true.append(1)  # Assume the test data contains relevant items
        y_pred.append(1 if anime_title in recommendations.values else 0)
    
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    
    return precision, recall, f1

# Example evaluation
precision, recall, f1 = evaluate_recommendations(test_data, similarity_matrix)
print(f"Precision: {precision}, Recall: {recall}, F1-Score: {f1}")
```
